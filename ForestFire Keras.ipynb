```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

from keras.models import Sequential
from keras.initializers import Initializer
from keras.layers import Dense
from keras.utils import to_categorical
from keras.callbacks import Callback
```

```python
setwd = 'C:\\Users\\PraveenPC\\Desktop\\Zekelabs\\Projects\\ForestFires\\'
rawfile = pd.read_csv(setwd + 'forestfires.csv')
SEED = 1000
```

```python
le = LabelEncoder()
rawfile['month'] = le.fit_transform(rawfile.month)
rawfile['day'] = le.fit_transform(rawfile.day)
```

```python
feature_cols = list(set(rawfile.columns) - set(['area']))
finalraw = rawfile[feature_cols]
finalrawtarget = rawfile['area']
#finalrawtarget
```

```python
#Both objects needs to convert to matrix
#finalraw = finalraw.as_matrix()
#finalrawtarget = finalrawtarget.as_matrix()
```

```python
X_train, X_test, y_train, y_test = train_test_split(finalraw, finalrawtarget, test_size=0.2, random_state=SEED)
```

```python
model = Sequential()
model.add(Dense(30,input_dim = X_train.shape[1], activation = 'sigmoid'))
model.add(Dense(20,activation = 'sigmoid'))
#model.add(Dense(25,activation='relu'))
model.add(Dense(1))

model.compile(loss="mean_squared_error", optimizer = 'adam', metrics = ['accuracy'])
```

```python
class history_loss(Callback):
    def on_train_begin(self, logs={}):
        self.losses = []

    def on_batch_end(self, batch, logs={}):
        batch_loss = logs.get('loss')
        self.losses.append(batch_loss)
```

```python
n_epochs = 500
batch_size = 50
validation_split = 0.2
history_relu = history_loss()
model.fit(X_train, y_train, epochs=n_epochs, batch_size=batch_size, callbacks=[history_relu], 
          validation_split=validation_split, verbose=2)
```
